```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("student-mat.csv", sep=';')
data.head()
```

```python
data.info()
```

```python
data = data.drop_duplicates()
```

```python
data['G3'].mean()
```

```python
(data['G3'] > 15).sum()
```

```python
data[['studytime', 'G3']].corr()
```

```python
data.groupby('sex')['G3'].mean()
```

```python
plt.hist(data['G3'], bins=10, edgecolor='black')
plt.title("Final Grades Distribution")
plt.xlabel("Grade")
plt.ylabel("Number of Students")
plt.show()
```

```python
plt.scatter(data['studytime'], data['G3'])
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.show()
```

```python
data.groupby('sex')['G3'].mean().plot(kind='bar')
plt.title("Average Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Grade")
plt.show()
```
